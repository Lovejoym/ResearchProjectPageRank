### PageRank analysis on social network- A case of AIMS 2021 students network



Corona virus is a virus that causes infections in nose or upperthroat. Different variants of the virus have been spreading worldwide. The main spread of the virus is from person to person. Transmission occurs when someone who is infected talks, coughs or sneezes and droplets carrying the virus enter the air. Then if someone close breathes in the infected air, they also become infected. The most influential individual in a network would be most likely to spread the virus to a large population if it happens that they are infected. Finding the most influential individuals would be important in monitoring and reducing the spread of the virus. Decisions such as limiting the movement of most influential individuals might be put in place so as to reduce the probability of these individuals from getting the virus and spreading it to others.

In this research we use PageRank to find the most influential individual in a network who is the individual most likely to spread the virus more than others. We collected a dataset from AIMS students, the 2020-21 intake and the 2021 intake. Students were asked whom they interact with more often. A directed network was created using the dataset. Analysis is done using different teleportation factors and types of PageRank.

### Using Reverse Pagerank to find the most influencial individual.

```python
def reverse(Y,alpha):
    """
    function returns the pagerank of nodes using the power method
    """
    A=nx.to_numpy_array(Y)#adjacency matrix of a graph
    nodelist= list(Y.nodes)
    S=np.nan_to_num(A/A.sum(axis=0))
    #S=substochastic(Y)
    l=len(Y)
    E=(1/l)*(np.ones((l,l)))
    M=(alpha*S)+((1-alpha)*E)#google matrix
    V = np.ones(l)/l#initial v
    for i in range(50):
        V = np.dot(M, V)
    V1=list(V)
    pageranks=dict(zip(nodelist,V1))
    key_list = list(pageranks.keys())
    val_list = list(pageranks.values())

    pageranks_sort = {}

    sort_val = sorted(val_list, reverse=True)

    for i,j in enumerate(sort_val):
        position = val_list.index(j)
        pageranks_sort[key_list[position]] = i+1
    return dict(sorted(pageranks_sort.items()))
    #return V1
```

### Using Weak Preferential PageRank on AIMS network

```python
def substochastic(Y):
    """
    The function returns a substochatsic matrix of a particular graph
    """
    A=nx.to_numpy_array(Y)#adjacency matrix of a graph
    B=A.transpose()
    d=B.sum(axis=0)
    Dp=np.linalg.pinv(np.diag(d))
    S=np.dot(B,Dp)
    return (S)
```

```python
def weakpreferential(Y,alpha=0.85):
    S=substochastic(Y)
    l=len(Y)
    nodelist= list(Y.nodes)
    #creating correction vector
    c=S.sum(axis=0) #create vector of the column sums of S
    for i in range (len(c)):
        if c[i]==0:
            c[i]=1 #for dangling node, sum=0
        else:
            c[i]=0 
            
    u=np.full((l,1),1/l) #uniform distibution vector
    S1=S+(u*c)   #stochastic matrix 
    E=(1/l)*(np.ones((l,l)))
    M=(alpha*S1)+((1-alpha)*E)#google matrix
    V = np.ones(l)/l#initial v
    for i in range(50):
        V = np.dot(M, V)
    V1=list(V)
    pageranks=dict(zip(nodelist,V1))
    key_list = list(pageranks.keys())
    val_list = list(pageranks.values())

    pageranks_sort = {}

    sort_val = sorted(val_list, reverse=True)

    for i,j in enumerate(sort_val):
        position = val_list.index(j)
        pageranks_sort[key_list[position]] = i+1
    return dict(sorted(pageranks_sort.items()))
    
```

### Using Weighted PageRank on AIMS Network

```python
from numpy.linalg import multi_dot
def weightedpagerank(Y, alpha=0.85):
    A=nx.to_numpy_array(Y)#adjacency matrix of a graph
    B=A.transpose() 
    d=A.sum(axis=0)+ A.sum(axis=1)#sum of in and out degree
    D=np.diag(d) #creating matrix D
    l=len(Y)
    nodelist= list(Y.nodes)

    e=np.ones((l,1))
    b=multi_dot([A,D,e]).flatten()
    p=np.linalg.pinv(np.diag(b))
    #creating substochastic matrix
    S=multi_dot([D,B,p])
    E=(1/l)*(np.ones((l,l)))
    M=(alpha*S)+((1-alpha)*E)#google matrix
    V = np.ones(l)/l#initial v
    for i in range(50):
        V = np.dot(M, V)
    V1=list(V)
    pageranks=dict(zip(nodelist,V1))
    key_list = list(pageranks.keys())
    val_list = list(pageranks.values())

    pageranks_sort = {}

    sort_val = sorted(val_list, reverse=True)

    for i,j in enumerate(sort_val):
        position = val_list.index(j)
        pageranks_sort[key_list[position]] = i+1
    return dict(sorted(pageranks_sort.items()))
    #return(V1)
```

### Using Strongly Preferential PageRank on AIMS Network

```python
def strongpreferential(Y,alpha=0.85):
    S=substochastic(Y)
    l=len(Y)
    nodelist= list(Y.nodes)

    #creating correction vector
    c=S.sum(axis=0) #create vector of the column sums of S
    for i in range (len(c)):
        if c[i]==0:
            c[i]=1 #for dangling node, sum=0
        else:
            c[i]=0 
    #Creating teleportation distribution        
    v=np.full((l,1),0)
    v[3,0]=1 # teleports to node 2 in teleportation
    #stochastic matrix 
    S1=S+(v*c)   
    E=(1/l)*(np.ones((l,l)))
    M=(alpha*S1)+((1-alpha)*E)#google matrix
    V = np.ones(l)/l#initial v
    for i in range(50):
        V = np.dot(M, V)
    V1=list(V)
    pageranks=dict(zip(nodelist,V1))
    key_list = list(pageranks.keys())
    val_list = list(pageranks.values())

    pageranks_sort = {}

    sort_val = sorted(val_list, reverse=True)

    for i,j in enumerate(sort_val):
        position = val_list.index(j)
        pageranks_sort[key_list[position]] = i+1
    return dict(sorted(pageranks_sort.items()))
    #return(V1)
```
